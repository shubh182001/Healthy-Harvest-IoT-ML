```python
import pandas as pd
croptobe=pd.read_csv("Crop_recommendation.csv",encoding ='latin1')
from sklearn import preprocessing
import numpy as np
croptobe.head()
```

```python
croptobe.shape
```

```python
classess = list(croptobe["label"].unique())
classess
```

```python
le = preprocessing.LabelEncoder()
croptobe['label']=le.fit_transform(croptobe['label'])
```

```python
import random
lst=[]
for i in range(2201):
    n = random.randint(25,65)
    lst.append(n)
dff=pd.DataFrame(lst,columns=['Moisture'])
```

```python
new_df=croptobe.join(dff)
```

```python
# dff_new=new_df[['N','P','K','Moisture','label']]
```

```python
x=new_df.drop(['label','rainfall'],axis=1)#independent
y=new_df['label']#dependent
```

```python
x_train,x_test , y_train, y_test = train_test_split(x, y,test_size=0.1)
```

```python
print("x_train shape: ",x_train.shape)
print("x_test shape: ",x_test.shape)
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

# Neural Network

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
```

```python
import tensorflow as tf
from tensorflow import keras
model=tf.keras.Sequential([
    keras.layers.Dense(20,input_shape=x_train.shape[1:],activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(22,activation='sigmoid'),
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=200, batch_size=1000)
```

```python
model.evaluate(x_test,y_test)
```

```python
pred=model.predict(x_test)
```

```python
y_pred_new=[]
for i in pred:
#     print(i)
    #break
    y=np.argmax(i)
    #break
    #print(y)
    y_pred_new.append(y)
    #break
y_pred_new
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred_new, target_names=classess))
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,y_pred_new)
import pandas as pd
df_cm=pd.DataFrame(cm)
import seaborn as sns
plt.figure(figsize = (15,7))
sns.heatmap(df_cm, annot=True)
```
